```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('HR_comma_sep.csv')
```

```python
data.head()
```

```python
X = data[['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident','promotion_last_5years','sales','salary']]
y = data['left']

from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
X['salary'] = le.fit_transform(X['salary'])
X['sales'] = le.fit_transform(X['sales'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=42)
```

```python
from  sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(6,5), random_state=5, verbose=True, learning_rate_init=0.01)
clf.fit(X_train, y_train)
```

```python
ypred = clf.predict(X_test)
ypred
```

```python
from sklearn.metrics import accuracy_score  
accuracy_score(y_test, ypred)
```
